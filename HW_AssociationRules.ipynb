```python
import pandas as pd

from mlxtend.frequent_patterns import apriori

from mlxtend.frequent_patterns import association_rules
```

```python
df= pd.read_excel("./CatalogCrossSell.xlsx", sheet_name="Data")
```

```python
df.head(5)
```

```python
#dropping the empty column
df.drop(df.columns[[1]], axis = 1, inplace = True) 
df.head(5)
```

```python
#checking for null values
df.isna().sum()
```

```python
df.drop("Customer Number", axis=1, inplace=True)
```

```python
#generating frequent items
frequent_itemsets = apriori(df, min_support=0.07, use_colnames=True)
frequent_itemsets
```

```python
#generating rules
rules = association_rules(frequent_itemsets, metric="lift", min_threshold=1)
rules.head()
```

```python
#sorting the rules in descending order to show the rules that has higher lift value
rules.sort_values(by="lift", ascending=False, inplace=True)
```
